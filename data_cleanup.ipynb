```python
# Import Dependencies
from pathlib import Path
import pandas as pd
```

```python
# Create a file path to the owid-covid-data.csv file.
csv_path = Path("owid-covid-data.csv")

# Import the owid-covid-data.csv file as a DataFrame
case_dataset_df = pd.read_csv(csv_path)

case_dataset_df.head()
```

```python
# Remove unnecessary columns from the DataFrame and save the new DataFrame

case_dataset_reduced_df = case_dataset_df[["location", "date", "total_cases_per_million", "new_cases_smoothed_per_million", "total_deaths_per_million", 
                       "new_deaths_smoothed_per_million", "icu_patients_per_million", 
                       "hosp_patients_per_million", "total_vaccinations_per_hundred", "people_fully_vaccinated_per_hundred",
                      "total_boosters_per_hundred", "new_vaccinations_smoothed_per_million", "new_people_vaccinated_smoothed_per_hundred",
                     "population_density", "median_age", "extreme_poverty", "cardiovasc_death_rate",
                      "diabetes_prevalence", "female_smokers", "male_smokers", "hospital_beds_per_thousand", "population"]]
case_dataset_reduced_df = case_dataset_reduced_df.rename(columns={"location": "CountryName",
                                        "date": "Date"})

case_dataset_reduced_df.head()
```

```python
# Filter to only keep countries in question

countries = ['Australia', 'India', 'Brazil', 'Vietnam']
case_dataset_reduced_df = case_dataset_reduced_df[case_dataset_reduced_df['CountryName'].isin(countries)]
case_dataset_reduced_df
```

```python
case_dataset_reduced_df['Date'] = pd.to_datetime(case_dataset_reduced_df['Date'], format='%d/%m/%Y').dt.strftime('%Y%m%d')
case_dataset_reduced_df
```

```python
case_dataset_reduced_df['Date'] = case_dataset_reduced_df['Date'].astype('int64')
```

```python
# Create a file path to the OxCGRT_compact_national.csv file.
csv_path_02 = Path("OxCGRT_compact_national.csv")

# Import the csv file as a DataFrame
policy_dataset_df = pd.read_csv(csv_path_02)

policy_dataset_df.head()
```

```python
# Remove unnecessary columns from the DataFrame and save the new DataFrame

policy_dataset_df = policy_dataset_df[["CountryName", "Date", "C1M_School closing", "C2M_Workplace closing", "C3M_Cancel public events", 
                       "C4M_Restrictions on gatherings", "C5M_Close public transport", 
                       "C6M_Stay at home requirements", "C7M_Restrictions on internal movement", "C8EV_International travel controls",
                      "E1_Income support", "E2_Debt/contract relief", "E3_Fiscal measures",
                     "E4_International support", "H1_Public information campaigns", "H2_Testing policy", "H3_Contact tracing",
                      "H4_Emergency investment in healthcare", "H5_Investment in vaccines", "H6M_Facial Coverings", "H8M_Protection of elderly people", 
                                       "StringencyIndex_Average", "GovernmentResponseIndex_Average", "ContainmentHealthIndex_Average", "EconomicSupportIndex"]]

policy_dataset_df
```

```python
policy_dataset_reduced_df = policy_dataset_df[policy_dataset_df['CountryName'].isin(countries)]

policy_dataset_reduced_df
```

```python

column_dtype = policy_dataset_reduced_df['Date'].dtype
column_dtype
```

```python
# Identify overlapping date range
start_date = max(case_dataset_reduced_df['Date'].min(), policy_dataset_reduced_df['Date'].min())
end_date = min(case_dataset_reduced_df['Date'].max(), policy_dataset_reduced_df['Date'].max())

# Filter dataframes based on overlapping date range
case_df_overlap = case_dataset_reduced_df[(case_dataset_reduced_df['Date'] >= start_date) & (case_dataset_reduced_df['Date'] <= end_date)]
policy_df_overlap = policy_dataset_reduced_df[(policy_dataset_reduced_df['Date'] >= start_date) & (policy_dataset_reduced_df['Date'] <= end_date)]

# Merge dataframes based on CountryName and Date
merged_df = pd.merge(case_df_overlap, policy_df_overlap, on=['CountryName', 'Date'])
merged_df
```

```python
# Push the remade DataFrame to a new CSV file
merged_df.to_csv("data_merge.csv", index=False, header=True)
```
